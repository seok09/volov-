```python
# 1. í•„ìˆ˜ íŒ¨í‚¤ì§€ ì„¤ì¹˜
!pip install ultralytics yt-dlp moviepy -q

# 2. ì„í¬íŠ¸
from ultralytics import YOLO
import yt_dlp
from moviepy.editor import VideoFileClip
import os
from IPython.display import Video, display

# 3. ìœ íŠœë¸Œ ì˜ìƒ ë‹¤ìš´ë¡œë“œ í•¨ìˆ˜
def download_youtube_video(url, output_name="input_video.mp4"):
    ydl_opts = {
        'format': 'bestvideo+bestaudio/best',
        'outtmpl': output_name,
        'merge_output_format': 'mp4',
        'quiet': True
    }
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(url, download=True)
        filename = ydl.prepare_filename(info)
        if not filename.endswith(".mp4"):
            filename = os.path.splitext(filename)[0] + ".mp4"
    if os.path.exists(filename):
        print(f"âœ… ë‹¤ìš´ë¡œë“œ ì™„ë£Œ: {filename}")
        return filename
    else:
        raise FileNotFoundError("âŒ ë‹¤ìš´ë¡œë“œëœ ì˜ìƒ íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")

# 4. ì˜ìƒ ìë¥´ê¸° (30ì´ˆ)
def trim_video(input_path, output_path="short_clip.mp4", duration=30):
    clip = VideoFileClip(input_path).subclip(0, duration)
    clip.write_videofile(output_path, codec="libx264", audio_codec="aac", verbose=False, logger=None)
    clip.close()
    print(f"âœ… {duration}ì´ˆ ì˜ìƒ ì €ì¥ ì™„ë£Œ: {output_path}")
    return output_path

# 5. YOLOv11 íƒì§€ í•¨ìˆ˜
def detect_objects(video_path, model_name="yolov11n.pt", conf=0.4, imgsz=640):
    print(f"ğŸ” {model_name} ëª¨ë¸ ë¡œë“œ ë° ê°ì²´ íƒì§€ ì‹œì‘...")
    model = YOLO(model_name)

    results = model.predict(
        source=video_path,
        conf=conf,
        imgsz=imgsz,
        save=True,       # ê²°ê³¼ ì €ì¥
        save_txt=True    # í…ìŠ¤íŠ¸ ë¼ë²¨ ì €ì¥
    )

    output_folder = results[0].save_dir
    print(f"ğŸ“‚ ê²°ê³¼ ì €ì¥ í´ë”: {output_folder}")

    # ê²°ê³¼ ì˜ìƒì€ ì¼ë°˜ì ìœ¼ë¡œ predictì—ì„œ ì…ë ¥ ì˜ìƒëª…ê³¼ ë™ì¼í•˜ê²Œ ì €ì¥ë¨
    output_video_path = os.path.join(output_folder, os.path.basename(video_path))

    # ì €ì¥ëœ íŒŒì¼ ì¡´ì¬ ì—¬ë¶€ í™•ì¸
    if os.path.exists(output_video_path):
        print(f"âœ… ê°ì²´ íƒì§€ ì™„ë£Œ! ê²°ê³¼ ì˜ìƒ ìœ„ì¹˜: {output_video_path}")
    else:
        # íŒŒì¼ëª… ë‹¤ë¥¼ ìˆ˜ ìˆìœ¼ë‹ˆ í´ë” ë‚´ íŒŒì¼ ë¦¬ìŠ¤íŠ¸ ì¶œë ¥
        print("âš ï¸ ê²°ê³¼ ì˜ìƒì´ ì˜ˆìƒ ìœ„ì¹˜ì— ì—†ìŠµë‹ˆë‹¤. í´ë” ë‚´ íŒŒì¼ ëª©ë¡:")
        print(os.listdir(output_folder))

    return output_video_path

# 6. ê²°ê³¼ ì˜ìƒ ì¶œë ¥ í•¨ìˆ˜
def show_video(video_path):
    if os.path.exists(video_path):
        display(Video(video_path, embed=True))
    else:
        print("âŒ ê²°ê³¼ ì˜ìƒ íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")

# 7. ì „ì²´ ì‹¤í–‰ í•¨ìˆ˜
def run_yolo_on_youtube(youtube_url):
    try:
        # ë‹¤ìš´ë¡œë“œ
        original_video = download_youtube_video(youtube_url)
        # 30ì´ˆ ìë¥´ê¸°
        trimmed_video = trim_video(original_video, duration=30)
        # íƒì§€
        result_video = detect_objects(trimmed_video)
        # ê²°ê³¼ ì¶œë ¥
        show_video(result_video)
    except Exception as e:
        print("ğŸš¨ ì˜¤ë¥˜ ë°œìƒ:", e)

# 8. ì‹¤í–‰ (URLë§Œ ë°”ê¾¸ë©´ ë¨)
youtube_url = "https://www.youtube.com/shorts/3op_P8DH4Rs"
run_yolo_on_youtube(youtube_url)
```
